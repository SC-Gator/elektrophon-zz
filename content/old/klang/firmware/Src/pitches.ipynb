```python
def writeArray(table = [], name="wavetable"):   
  print("\uint32_t %s[%d] = {\n\t" %(name,len(table)), end='')
  rows = 0
  for i in range(1,len(table)):
    if rows > 10 :
      rows = 0
      print("\n\t", end='')
    print("%4.0d, " %(table[i]), end = '')
    rows = rows + 1
  print("\n};")

print("/* autogenerated wavetables */")
print("#include \"wavetable.h\"")

x = np.linspace(-np.pi, np.pi, 1024)
data = np.sin(x)
plot.plot(x, (data+1)*(4096-1)/2)
writeArray((data+1)*(4096-1)/2, "wt_sine")

```
